### Importación de librería

```python
import pandas as pd
import numpy as np
```

### Cargando datos del archivo maestro

```python
train = pd.read_csv('C:\\Users\\Usuario\\Desktop\\MachineLearningTest-master\\data\\train.csv')
```

### Cabecera de los datos

```python
train.head(6)
```

### Eliminación de los NaN

```python
train = train.fillna(train.mean())
```

```python
train
```

```python
for i, val in enumerate(train.num):
    if val != 0:
        train.num[i] = 1 # train.loc[i,'num'] = 1
        #DF.loc['juan','edad'] #escribir
        #DF.iloc[5,3]          #escribir entero
```

```python
train.head(5)
```

### Definir matrices para entrenar (todas menos col num) y predecir (col num)

```python
X, y = train.iloc[:,:-1], train.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=38)
```

```python
display(X_train) #X_train.head()
```

### Regresión logística

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
y_pred = logistic.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision_recall_curve(y_test, y_pred)
```

```python
from sklearn.metrics import log_loss
```

```python
y_prob = logistic.fit(X_train,y_train).predict_proba(X_test)
```

```python
log_loss(y_test, y_prob[:,1])
```

### Probar árboles de classificación

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arbol = DecisionTreeClassifier()
```

```python
y_pred = arbol.fit(X_train, y_train).predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
y_prob = arbol.fit(X_train,y_train).predict_proba(X_test)
```

```python
y_prob
```

```python
log_loss(y_test, y_prob[:,1])
```
